```python
import numpy as np
from scipy import linalg
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

для многочлена первой степени

```python
A = np.array([[1., 1.], [1., 15.]]) # Матрица (левая часть системы)
b = np.array([[f(1)], [f(15)]])
coefficient =  linalg.solve(A, b)
print(coefficient)
```

```python
x = np.arange(1, 15, 0.1)
y = f(x)
plt.plot(x, y)
x1 = np.array([1, 15])
y1 = coefficient[0] + coefficient[1] * x1
plt.plot(x1, y1)
plt.show()
```

Многочлен второй степени

```python
A = np.array([[1., 1., 1.], [1., 8., 64.], [1., 15., 225]]) # Матрица (левая часть системы)
b = np.array([[f(1)],[f(8)], [f(15)]])
coefficient =  linalg.solve(A, b)
print(coefficient)
```

```python
x = np.arange(1, 15, 0.1)
y = f(x)
plt.plot(x, y)
x1 = np.array([1, 8, 15])
y1 = coefficient[0] + coefficient[1] * x1 + coefficient[2] * x1**2
plt.plot(x1, y1)
plt.show()
```

Многочлен третьей степени

```python
A = np.array([[1., 1., 1., 1.], [1., 4., 16., 64.], [1., 10., 100., 1000], [1., 15., 225, 3375]]) # Матрица (левая часть системы)
b = np.array([[f(1)],[f(4)], [f(10)], [f(15)]])
coefficient =  linalg.solve(A, b)
print(coefficient)
```

```python
x = np.arange(1, 15, 0.1)
y = f(x)
plt.plot(x, y)
x1 = np.array([1, 4, 10, 15])
y1 = coefficient[0] + coefficient[1] * x1 + coefficient[2] * x1**2 + coefficient[3] * x1 ** 3
plt.plot(x1, y1)
plt.show()
```
